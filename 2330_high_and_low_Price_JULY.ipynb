```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv( './drive/MyDrive/STOCK_DAY_2330_202206.csv', encoding='utf-8')
```

```python
data.head(22)
```

```python
left = np.array([1,2,6,7,8,9,10,13,14,15,16,17,20,21,22,23,24,27,28,29,30])
height = data.iloc[1:22, 4:5]
h = np.array(height)
int_list = list(map(float,h))
print(int_list)
print(left)
```

```python
plt.ylim(470,560)
plt.xlim(1,30)
plt.bar(left,int_list,width=1.0,color = 'red')
plt.show()
```

```python
height2 = data.iloc[1:22, 5:6]
h2 = np.array(height2)
int_list2 = list(map(float,h2))
plt.ylim(470,560)
plt.xlim(1,30)
plt.bar(left,int_list,width=1.0,color = 'red')
plt.bar(left,int_list2,width=1.0,color = 'green')
plt.show()
```

```python
plt.title("TPE:2330 July", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.xlabel("Day", fontdict={'fontname': 'Arial'})
plt.ylabel("Price", fontdict={'fontname': 'Arial'})
plt.ylim(470,560)
plt.xlim(1,30)
plt.bar(left,int_list,width=1.0,color = 'red')
plt.bar(left,int_list2,width=1.0,color = 'green')
plt.show()
```
